```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
label_encoder = LabelEncoder()
```

```python
data['chest pain type'] =label_encoder.fit_transform(data['chest pain type'])
```

```python
data['ST slope'] = label_encoder.fit_transform(data['ST slope'])
```

```python
data.head()
```

```python
x = data.drop('target',axis=1)
```

```python
y = data['target']
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled_df = pd.DataFrame(x_scaled , columns = x.columns)
```

```python
x_scaled_df
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))

# Plotting histograms and KDEs for each feature in X_scaled_df
for i, column in enumerate(x_scaled_df.columns):
    plt.subplot(3, 4, i+1)  # Adjust the layout as needed, here it's a 3x4 grid
    sns.histplot(x_scaled_df[column], kde=True, bins=30)
    plt.title(f"Distribution of {column}")

# Adjust the layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_scaled_df,y , test_size = 0.2,random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(Y_test,y_pred))
print("Classification Report:")
print(classification_report(Y_test , y_pred))
```
